## Compute co-expression scores and gene co-activity scores (from squared gene-gene and peak-gene correlations)


#### Gene co-expression scores

```python
### Specify input files
coexpression_file = "" ### Fill in name of file with gene-gene correlations
coexpression_scores_outfile = "" ### Fill in name of file to save gene co-expression scores to
background_coexpression_file = "" ### Fill in name of file with expected upward bias in squared gene-gene correlations
###

if not os.path.exists(coexpression_scores_outfile):
    print("Computing coexpression scores!")
    coexpression = pd.read_csv(coexpression_file, sep = '\t')
    coexpression = fn.square_column(coexpression, 'correlation')
    
    # Correct each r2
    background_coexpression = pd.read_csv(background_coexpression_file, sep = '\t')['noise'].iloc[0]
    coexpression['correlation_squared'] = coexpression['correlation_squared'] - background_coexpression
    coexpression['cis_genes'] = 1

    # Compute scores
    coexpression_scores = coexpression[['gene', 'correlation_squared', 'cis_genes']].groupby('gene').agg(sum).reset_index(
        ).rename(columns = {'correlation_squared': 'coexpression_score'})
    coexpression_scores.to_csv(coexpression_scores_outfile, sep = '\t', index = False)
```

#### Gene co-activity scores

```python
### Specify input files
peak_gene_links_file = "" ### Fill in name of file with peak-gene links
gene_coactivity_scores_outfile = "" ### Fill in name of file to save gene co-activity scores to
background_coactivity_file = "" ### Fill in name of file with expected upward bias in squared peak-gene correlations
###

if not os.path.exists(gene_coactivity_scores_outfile):
    print("Computing gene coactivity scores!")
    coactivity = pd.read_csv(peak_gene_links_file, sep = '\t')
    coactivity = fn.square_column(coactivity, 'correlation')

    # Correct r2
    background_coactivity = pd.read_csv(background_coactivity_file, sep = '\t')['noise'].iloc[0]
    coactivity['correlation_squared'] = coactivity['correlation_squared'] - background_coactivity
    coactivity['cis_peaks'] = 1

    # Compute scores
    coactivity_scores = coactivity[['gene', 'correlation_squared', 'cis_peaks']].groupby('gene').agg(sum).reset_index(
        ).rename(columns = {'correlation_squared': 'coactivity_score'})
    coactivity_scores.to_csv(gene_coactivity_scores_outfile, sep = '\t', index = False)
```
