## Compute co-expression scores (from squared gene-gene and peak-gene correlations)


```python
### Import packages

import pandas as pd
import numpy as np
import os
import sys

sys.path.insert(1, '../')
import functions as fn
```

#### Gene co-expression scores

```python
### Specify input files
coexpression_file = "../example_outputs/gene_scores/gene_gene_coexpression.tsv" ### Fill in name of file with gene-gene correlations
coexpression_scores_outfile = "../example_outputs/gene_scores/gene_coexpression_scores.tsv" ### Fill in name of file to save gene co-expression scores to
background_coexpression_file = "../example_outputs/gene_scores/background_coexpression.tsv" ### Fill in name of file with expected upward bias in squared gene-gene correlations
###

if not os.path.exists(coexpression_scores_outfile):
    print("Computing coexpression scores!")
    coexpression = pd.read_csv(coexpression_file, sep = '\t')
    coexpression = fn.square_column(coexpression, 'correlation')
    
    # Correct each r2
    background_coexpression = pd.read_csv(background_coexpression_file, sep = '\t')['noise'].iloc[0]
    coexpression['correlation_squared'] = coexpression['correlation_squared'] - background_coexpression
    coexpression['cis_genes'] = 1

    # Compute scores
    coexpression_scores = coexpression[['gene1', 'correlation_squared', 'cis_genes']].groupby('gene1').agg(sum).reset_index(
        ).rename(columns = {'correlation_squared': 'coexpression_score', 'gene1': 'gene'})
    coexpression_scores.to_csv(coexpression_scores_outfile, sep = '\t', index = False)
```
